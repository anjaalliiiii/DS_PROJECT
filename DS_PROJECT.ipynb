```python
x = data.drop(['Chance of Admit '],axis=1)
y= data ['Chance of Admit ']
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
sc_x.fit (x)
x_scaled = sc_x.transform (x)
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split (x_scaled, y, test_size=0.30, random_state=100)
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit (x_train, y_train)
LinearRegression()
y_pred = lin_reg.predict (x_test)
sns.regplot (x= y_test, y= y_pred)
```

```python
sns.distplot (y_test-y_pred)
```

```python
R_squared = lin_reg.score (x_train, y_train)
R_squared
```

```python
n = x.shape[0]
p = x.shape [1]
Adjusted_R_Squared = 1 - (1 - R_squared) * (n - 1)/(n - p -1)
Adjusted_R_Squared
```

```python
Weight =pd.DataFrame (lin_reg.coef_)
# Weight
from sklearn.feature_selection import f_regression
p_values = pd.DataFrame(f_regression (x_train, y_train)[1])
# p_values
Summary = pd.DataFrame (x.columns, columns=['Features'])
Summary['Weight'] = Weight
Summary['P_values'] = p_values
Summary
```
